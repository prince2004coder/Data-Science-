```python
import numpy as np


```

```python

matrix = np.random.randint(1, 91, (5, 5))

print(matrix)

column_mean =  np.mean(matrix, axis=0)
print("Column-wise Mean:",  column_mean)


row_mean  = np.mean(matrix, axis=1)
print("Row-wise Mean:",  row_mean)


matrix2 = np.where(matrix % 2 == 0, -1, matrix)
print(matrix2)

row_sorted = np.sort(matrix, axis=1)
print(row_sorted)


column_sorted = np.sort(matrix, axis=0)
print(column_sorted)



row_sum = np.sum(matrix, axis=1)
print(" Row wise sum:" ,row_sum)


column_sum = np.sum(matrix, axis=0)
print("Column wise sum:", column_sum)


count_column = np.sum(np.where(matrix2 < 0, 1, 0) , axis=0)
print("count colummn:", count_column)

```

```python
#question 2
#inventory control system

#tasks

stock = np.random.randint(20,80,(5,5))
print(stock)
demand = np.random.randint(20,80,(5,5))
print(demand)

remaning_stock = stock - demand
print(remaning_stock)

remaning_stock = np.where(remaning_stock < 0, 0, remaning_stock)
print(remaning_stock)


# find total shortage,,,,products with zero stock ,,,, sort each row based on remaining stock

total_shortage = np.sum(np.where(remaning_stock<0, remaning_stock*-1, 0))

print(total_shortage)


product_shortage = np.sum(np.where(remaning_stock<0, 1, 0), axis=0)
print(product_shortage)



sorted = np.sort(remaning_stock, axis=1)
print(sorted)
```
